```python
# Importing functions for performance evaluation: confusion matrix and classification report
from sklearn.metrics import confusion_matrix,classification_report
```

```python
# Creating a list of actual labels with 50 positive and 450 negative examples
y_act = [1]*50 + [0] * 450
# Creating a list of predicted labels with 45 true positives, 400 true negatives, and 55 false positives
y_pred = [1]*45 + [0] * 400 + [1] * 55
```

```python
print(confusion_matrix(y_act,y_pred))
```

```python
print(classification_report(y_act,y_pred))
```
